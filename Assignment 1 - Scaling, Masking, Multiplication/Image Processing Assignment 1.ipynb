### Qunatization

Start with your favorite image. Use the “shift right” operator and force the MSB position to zero. This will reduce the dynamic range to half (for example, 0-255 to 0-127, 0-64,0-32, 0-16, 0-8, 0-4, and 0-2). Explain the effect of quantization and compute the SNR.

#### Alogrithm

Shift right algorithm <br>
x >> y, return x with bits shifted by y places




```python
def ShiftPixels(image, step):
    # for each channel, do shiftwise operation on the image plane, pixel by pixel
    for i in range(image.shape[2]):
        for j in range(image.shape[1]):
            for k in range(image.shape[0]):
                image[k,j,i] =  image[k,j,i] >> step
    return image
```

```python
###################################################################
# Import Image

img = cv2.imread("Data/laptop.jpg")
# img = cv2.imread("Data/lap.jpg")
#img = cv2.imread("Data/m.jpg")
    
# Displaying the image
# cv2.imshow('image', img)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

###################################################################
# Resize the image by half in iteration and display the results 

row, col = 2, 4

f = plt.figure(figsize=(14, 8))

# Display the original Image
temp = img
f.add_subplot(row, col, 1)
plt.imshow(temp)
plt.title(256)

v= []

for i in range (row*col-1):
    k = f.add_subplot(row, col, i+2)
    temp = ShiftPixels(temp,1)
    k.title.set_text( 2**(7-i) )
    #plt.imshow(np.multiply(temp, 255))
    r = 255/(2**(7-i) - 1)
    v.append(r)
    plt.imshow(np.multiply(temp, r))
    #plt.imshow(np.rint (np.multiply(temp, 255/(2**(7-i) - 1))) )
```
